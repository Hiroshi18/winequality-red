```python
# Pre processamento dos dados
bins = (2, 6.5, 8)
groups = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins = bins, labels=groups)
```

```python
df
```

```python
df['quality'] = preprocessing.LabelEncoder().fit_transform(df['quality'])
```
